From the above we can observe, for features 'R','H','CG','SV','E' skewenwss values are not in range. Let's use power transformation techniques to reduce skeweness.

```python
from scipy.stats import boxcox
df['R']=np.log(df['R'])
df['E']=boxcox(df['E'])[0]
df['H']=np.cbrt(df['H'])
df['CG']=np.sqrt(df['CG'])
df['SV']=np.cbrt(df['SV'])
df.skew()
```

Let's check for outliers present in the dataset

```python
n=1
plt.figure(figsize=(20,20))
for i in df.columns:
    ax=plt.subplot(5,4,n)
    sns.boxplot(y=df[i])
    plt.xlabel(i)
    n+=1
    
plt.show()    
```

From the above we can observe outliers are present in features 'R','ERA','SHO','E'. Let's remove outliers with zscore technique.

```python
from scipy.stats import zscore

outlier_features=[ 'R','ERA','SHO','E']

zscore=np.abs(zscore(df[outlier_features]))

zscore>3
```

```python
df1=df[(zscore<3).all(axis=1)]
df1.shape
```

Let's check is there is multi collinearity isssue exists in the dataset.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
```

```python
x=df1.drop(columns='W',axis=1)
y=df1['W']
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
VIF=pd.DataFrame()
VIF['vif']=[variance_inflation_factor(x_scaled,i) for i in range(x.shape[1])]
VIF['features']=x.columns
VIF
```

Variance Inflation Factor for features 'RA','ER','ERA' are very high. Multi collinearity problem exists. 

As, per my observation from heatmap, 'R','ER','ERA' features are highly correalted with eachother.

'ER','ERA' features correlation value is 1.00, which means they are 100% correlated with each other.

Let's drop feature 'ER' because 'ERA' feature is contributing more for predicting target variable. 

```python
x.drop(columns=['ER'] ,axis=1, inplace=True)
```

```python
x_scaled=scaler.fit_transform(x)
```

Let's find the best Random State value.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
max_rvalue=0
max_randstate=0
for i in range(1,300):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=i)
    LR=LinearRegression()
    LR.fit(x_train,y_train)
    y_predict=LR.predict(x_test)
    r_value=r2_score(y_test,y_predict)
    if(r_value>max_rvalue):
        max_rvalue=r_value
        max_randstate=i
        
    
print('Best random state : ',max_randstate) 
print('r2 score :',max_rvalue)
```

Best Random State is 30 and R2score is 90%.

Let's build the model.

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=104)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor,ExtraTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
y_predict_LR=LR.predict(x_test)
r_value_LR=r2_score(y_test,y_predict_LR)
print("R2 score ",r_value_LR)
```

```python
L=Lasso()
L.fit(x_train,y_train)
y_predict_L=L.predict(x_test)
r_value_L=r2_score(y_test,y_predict_L)
print("R2 score ",r_value_L)
```

```python
R=Ridge()
R.fit(x_train,y_train)
y_predict_R=R.predict(x_test)
r_value_R=r2_score(y_test,y_predict_R)
print('R2 score ',r_value_R)
```

```python
DTR=DecisionTreeRegressor()
DTR.fit(x_train,y_train)
y_predict_DTR=DTR.predict(x_test)
r_value_DTR=r2_score(y_test,y_predict_DTR)
print('R2 score ',r_value_DTR)
```

```python
KNR=KNeighborsRegressor()
KNR.fit(x_train,y_train)
y_predict_KNR=KNR.predict(x_test)
r_value_KNR=r2_score(y_test,y_predict_KNR)
print('R2 score  ',r_value_KNR)
```

```python
RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)
y_predict_RFR=RFR.predict(x_test)
r_value_RFR=r2_score(y_test,y_predict_RFR)
print('R2 score ',r_value_RFR)
```

```python
ETR=ExtraTreeRegressor()
ETR.fit(x_train,y_train)
y_predict_ETR=ETR.predict(x_test)
r_value_ETR=r2_score(y_test,y_predict_ETR)
print('R2 score ',r_value_ETR)
```

From the above we can observe Ridge model provides highest R2 score of 86.5%. 
Let's consider this model as best model for this dataset.
