Outliers are present in Standard Error, Family,Trust,Generosity,Dystopia Residual. Let's remove the outliers with the help of Zscore technique

```python
from scipy.stats import zscore
cols=['Standard Error','Family','Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']
score=np.abs(zscore(df[cols]))
```

```python
np.where(score>3)
```

```python
df=df[(score<3).all(axis=1)]
df.shape
```

Let's use variance inflation factor for checking Multi collinearity issue

```python
x=df.drop(columns='Happiness Score')
y=df['Happiness Score']
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
vif_value=pd.DataFrame()
vif_value['vif']=[variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif_value['features']=x.columns
vif_value
```

Vif value of Happiness Rank feature is very high. Let's drop that feature.

```python
x.drop(columns='Happiness Rank',axis=1,inplace=True)
x.shape
```

```python
x_scaled=scaler.fit_transform(x)
```

Let's again check vif value

```python
vif_value=pd.DataFrame()
vif_value['vif']=[variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif_value['features']=x.columns
vif_value
```

After deleting Happiness Rank feature vif values of remaining values are also changed

Now let's find the best RandomState value

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
best_randstate=0
best_rvalue=0

for i in range(0,400):
    LR=LinearRegression()
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=i)
    LR.fit(x_train,y_train)
    y_predict=LR.predict(x_test)
    r_value=r2_score(y_test,y_predict)
    if(r_value>best_rvalue):
        best_rvalue=r_value
        best_randstate=i
        
        
print('Best Random State-',best_randstate,'\nR2 score -',best_rvalue)       
```

Let's split the data and build and train the model

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=370)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso,Ridge
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
y_predict_LR=LR.predict(x_test)
r_value_LR=r2_score(y_test,y_predict_LR)
print('R2 Score-',r_value_LR)
```

```python
RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)
y_predict_RFR=RFR.predict(x_test)
r_value_RFR=r2_score(y_test,y_predict_RFR)
print('R2 Score is ',r_value_RFR)
```

```python
KNN=KNeighborsRegressor()
KNN.fit(x_train,y_train)
y_predict_KNN=KNN.predict(x_test)
r_value_KNN=r2_score(y_test,y_predict_KNN)
print('R2 Score is ',r_value_KNN)
```

```python
DT=DecisionTreeRegressor()
DT.fit(x_train,y_train)
y_predict_DT=DT.predict(x_test)
r_value_DT=r2_score(y_test,y_predict_DT)
print('R2 Score is ',r_value_DT)
```

```python
LS=Lasso()
LS.fit(x_train,y_train)
y_predict_LS=LS.predict(x_test)
r_value_LS=r2_score(y_test,y_predict_LS)
print("R2 Score ",r_value_LS)
```

```python
R=Ridge()
R.fit(x_train,y_train)
y_predict_R=R.predict(x_test)
r_value_R=r2_score(y_test,y_predict_R)
print('R2 Score ',r_value_R)
```

From the above we can observe, out of all the models Linear Regression and Ridge models are providing high R2score. Let's check whether the value is due to overfitting of the model

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(RFR,x,y)
print('score ',score)
print('Mean ',score.mean())
print('Difference is ',(r_value_RFR-score.mean()))
```

```python
score=cross_val_score(LR,x,y)
print('score ',score)
print('Mean ',score.mean())
print('Difference ',(r_value_LR-score.mean()))
```

```python
score=cross_val_score(KNN,x,y)
print('score ',score)
print('Mean ',score.mean())
print('Difference ',(r_value_KNN-score.mean()))
```

```python
score=cross_val_score(DT,x,y)
print('Score ',score)
print('Mean ',score.mean())
print("Difference ",r_value_DT-score.mean())
```

```python
score=cross_val_score(LS,x,y)
print('Score ',score)
print('Mean ',score.mean())
print('Difference ',r_value_LS-score.mean())
```

```python
score=cross_val_score(R,x,y)
print('Score ',score)
print('Mean ',score.mean())
print('Difference ',r_value_R-score.mean())
```

Cross validation score of Linear Regression model is less compared to remaining models. Let's consider Linear Regression as the best model for this dataset

Let's calaculate Mean absolute error,Mean squared error and Root Mean squared error

```python
msr=mean_squared_error(y_test,y_predict_LR)
print('Mean Squared Error',msr)
mae=mean_absolute_error(y_test,y_predict_LR)
print('Mean Absolute Error',mae)
print('Mean Squared Error',np.sqrt(msr))
```

Linear Regression Model provides R2 score of 99.99.So, there is no need of Hyper Parameter Tuning
