```python
df3=pd.DataFrame(df)
```

```python
numerical_cols=[]
categorical_cols=[]

for i in df3.columns:
    if(df3[i].dtypes=='object'):
        categorical_cols.append(i)
        
        
    else:
        numerical_cols.append(i)
```

```python

plt.figure(figsize=(10,15))
n=1
for i in numerical_cols:
    ax=plt.subplot(4,2,n)
    sns.distplot(x=df3[i])
    plt.xlabel(i)
    n+=1
    
    
plt.show()    
```

```python
df3.skew()
```

From the above we can observe all numerical columns skeweness values are not in range. Let's reduce skeweness using power transformation techniques.

Price range feature has categorical values. Let's not remove skeweness from this feature.

```python

df3['Country Code']=np.cbrt(df3['Country Code'])
df3['Country Code']=np.cbrt(df3['Country Code'])
df3['Country Code']=np.cbrt(df3['Country Code'])

df3['Votes']=np.cbrt(df3['Votes'])
df3['Votes']=np.sqrt(df3['Votes'])

df3.skew()
```

Skeweness value of 'Longitude','Latitude','Aggregate rating' features are increasing if power transformtion techniques are applied. Let's not transform the data of these features.

Skew value for 'Country Code' is not decreasing even if power transformation techniques are applied multiple times. 

Let's try to find outliers in the data.

```python

plt.figure(figsize=(10,15))
n=1
for i in numerical_cols:
    ax=plt.subplot(4,2,n)
    sns.boxplot(y=df3[i])
    plt.xlabel(i)
    n+=1
    
    
plt.show()    
```

From the above we an observe outliers in 'Country Code','Longitude','Latitude','Price range','Aggregate rating','Votes' features.

Let's find best Random State.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
max_rvalue=0
max_randstate=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=i)
    LR=LinearRegression()
    LR.fit(x_train,y_train)
    y_predict=LR.predict(x_test)
    r_value=r2_score(y_test,y_predict)
    if(r_value>max_rvalue):
        max_rvalue=r_value
        max_randstate=i
        
    
print('Best random state : ',max_randstate) 
print('r2 score :',max_rvalue)
```
