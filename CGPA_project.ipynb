From the above I am unable to conclude whether there is multi collinearirt issue. Let's check this with the help of variance inflation factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
x=df.drop(columns='CGPA',axis=2)
y=df['CGPA']
```

```python
Vif=pd.DataFrame()
Vif['features']=x.columns
Vif['value']=[variance_inflation_factor(x,i) for i in range(x.shape[1])]
Vif
```

From the above I can observe there is no multi collinearity issue. Let's continue finding random state for building model.

Let's find the best Random state for the dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python

max_r_value=0
random_state=0
for i in range(1,400):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    LR=LinearRegression()
    LR.fit(x_train,y_train)
    y_predict=LR.predict(x_test)
    rvalue=r2_score(y_test,y_predict)
    if(rvalue>max_r_value):
        max_r_value=rvalue
        random_state=i
        
print('Best Random state-',random_state,'\n R2 score-',max_r_value)
```

Best Random State is 72 and R2 score is 96.4

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=72)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso,Ridge
from sklearn.ensemble import RandomForestRegressor
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
y_predict_LR=LR.predict(x_test)
r_value_LR=r2_score(y_test,y_predict_LR)
print("R2 score - ",r_value_LR)
```

```python
DTR=DecisionTreeRegressor()
DTR.fit(x_train,y_train)
y_predict_DTR=DTR.predict(x_test)
r_value_DTR=r2_score(y_test,y_predict_DTR)
print("R2 score - ",r_value_DTR)
```

```python
KNR=KNeighborsRegressor()
KNR.fit(x_train,y_train)
y_predict_KNR=KNR.predict(x_test)
r_value_KNR=r2_score(y_test,y_predict_KNR)
print('R2 score - ',r_value_KNR)
```

```python
L=Lasso()
L.fit(x_train,y_train)
y_predict_L=L.predict(x_test)
r_value_L=r2_score(y_test,y_predict_L)
print('R2 score - ',r_value_L)
```

```python
R=Ridge()
R.fit(x_train,y_train)
y_predict_R=R.predict(x_test)
r_value_R=r2_score(y_test,y_predict_R)
print('R2 score - ',r_value_R)
```

```python
RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)
y_predict_RFR=RFR.predict(x_test)
r_value_RFR=r2_score(y_test,y_predict_RFR)
print('R2 score - ',r_value_RFR)
```

From the above we can observe KNeighborsRegressor is giving highest r2 score of 96.7% than other models.Let's check whether score is due to overfitting of the model with the help of cross validation score

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(LR,x,y)
print("Difference is ",r_value_LR-score.mean())
```

```python
score=cross_val_score(DTR,x,y)
print("Difference is ",r_value_DTR-score.mean())
```

```python
score=cross_val_score(KNR,x,y)
print("Difference is ",r_value_KNR-score.mean())
```

```python
score=cross_val_score(L,x,y)
print("Difference is ",r_value_L-score.mean())
```

```python
score=cross_val_score(R,x,y)
print("Difference is ",r_value_R-score.mean())
```

```python
score=cross_val_score(RFR,x,y)
print("Difference is ",r_value_RFR-score.mean())
```

From the above we can observe RandomForestRegressor has less value when compared with other values.
Let's consider Random Forest Regressor as best model for the giver dataset.

R2 score for the given dataset is 94%. Let's find MAE,MSE and RMSE

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("Mean Absolute Error ",mean_absolute_error(y_test,y_predict))
print("Mean Squared Error ",mean_squared_error(y_test,y_predict))
print("Root mean squared error ",np.sqrt(mean_squared_error(y_test,y_predict)))
```
